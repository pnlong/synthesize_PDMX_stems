# Stable Audio 3 Preset Exploration

Prerequisites:
- `git submodule update --init synthesis/realify/stable-audio-3`
- SA3 installed in the submodule venv
- Sample stems under `{OUTPUT_DIR}/ablations/basic/` or `{OUTPUT_DIR}/stems/`

```python
import sys
from pathlib import Path

REPO = Path.cwd().parents[2]
SA3 = REPO / "synthesis" / "realify" / "stable-audio-3"
sys.path.insert(0, str(SA3))

import torchaudio
from stable_audio_3 import StableAudioModel

DATASET_DIR = Path("/deepfreeze/user_shares/pnlong/ablations/basic")
CAPTIONS_CSV = DATASET_DIR / "captions.csv"
```

```python
model = StableAudioModel.from_pretrained("medium")
STEM_PATH = DATASET_DIR / "data" / "..." / "stem_0.flac"
PROMPT = "format: solo | instruments: vibraphone, genres: jazz"

init_audio = torchaudio.load(str(STEM_PATH))
duration = init_audio[0].shape[-1] / init_audio[1]

for init_noise_level in [0.5, 0.65, 0.8, 0.9]:
    audio = model.generate(
        init_audio=init_audio,
        init_noise_level=init_noise_level,
        prompt=PROMPT,
        duration=duration,
    )
    out = Path(f"realify_out/noise_{init_noise_level}.flac")
    out.parent.mkdir(exist_ok=True)
    torchaudio.save(str(out), audio, sample_rate=44100)
    print(f"Wrote {out}")
```
